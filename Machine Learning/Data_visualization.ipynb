#Example 15
Use the **mtcars** dataset, and visualize **disp** in terms of **am** and  **gear**  using  an intractive  visualisation including a scatter plot, bar plot, and histogram in Plotly. 

```python
import plotly.graph_objects as px
import pandas as pd
 
# reading the database
data = pd.read_csv("mtcars.csv")
 
 
plot = px.Figure(data=[px.Scatter(
    x=data['cyl'],
    y=data['mpg'],
    mode='markers',)
])
 
# Add dropdown
plot.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="left",
            buttons=list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                ),
                dict(
                    args=["type", "histogram"],
                    label="Histogram",
                    method="restyle"
                )
            ]),
        ),
    ]
)
 
plot.show()
```

```python

# importing the modules
from bokeh.plotting import figure, output_file, show
import pandas as pd
 
 
# instantiating the figure object
graph = figure(title = "Bokeh Bar Chart")
 
# reading the database
data = pd.read_csv("mtcars.csv")
 
# plotting the graph
graph.vbar(data['am'], top=data['disp'],
           legend_label = "disp vs am", color='green')
 
graph.vbar(data['gear'], top=data['disp'],
           legend_label = "disp VS gear", color='red')
 
graph.legend.click_policy = "hide"
 
# displaying the model
show(graph)
```
