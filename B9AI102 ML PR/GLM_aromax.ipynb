## Genralized linear models (GLM): GLM is a set of supervised machine learning techniques to accomplish the task of predictive analytics

## Different types of tasks in predictive Analytics: 



1.   Regression analysis
2.   Classification problem

**Regression analysis**: the task is to predict the **continuous output variable**. 

**Classification problem**:  the task is to predict the **categorical output variable**. 


## supervised ML (SML): SML is an input/output system to design a predictive model. 

image.png

image.png

*  Input variables: independent variables, predictors, explantory varibles

* output varible: dependent variable, target variable
# Note: GLM is a set of techniques to solve SML. 



#Different types of GLM



1.   Linear regression (Regression model)
2.   Logistic regression (Classification model)

* binary logistic reg

*  multinomial logistic reg

3.   Poisson Regression (both)

**Key differences:**

1. The type of output variable
2. The link function: is a mathematical function to connect input variables to the output variable

e.g. 
1.   Linear regression (The type of output variable: **continuous**, link function is **identity** function);

 y=t

2.   Logistic regression (The type of output variable: **categorical**, link function is **logistic** function) 

 y=1/(1+exp(-t))

3.   Poisson Regression ((The type of output variable: **counts**, link function is **exponential** function))   

y=exp(-t)

````python
# statistical way to recall glm
import statsmodels.api as sm
from statsmodels.api import GLM, add_constant
from pandas import read_csv, DataFrame

dataset = read_csv('dataset/boston_house_prices.csv')

X = dataset.drop('TAX', axis=1)  # input variables
y = dataset['TAX']  # output variable
X = add_constant(X)
sf= ['const', 'ZN', 'INDUS', 'RAD', 'MEDV']
X = X[sf]

# create the linear reg model using
# model = GLM(y, X, family=sm.families.Gaussian(sm.families.links.log))       # deprecated
model = GLM(y, X, family=sm.families.Gaussian(sm.families.links.log()))
res = model.fit()
res.summary()
# pvalues = dict(res.pvalues)
# for x in pvalues:
#     if pvalues[x] < 0.05:
#         print(x)
````

````python
# create the linear regression model

from numpy import transpose
model= GLM(y, X, family=sm.families.Gaussian(sm.families.links.log)) # create the linear reg model using
#fit the model
trained_model = model.fit()
#res.summary()
w=round(trained_model.params,2)
pvalues=round(trained_model.pvalues,3)
df=transpose(DataFrame([w,pvalues]))
df.columns=['weights','pvalues']
df
````

````python
from pandas import read_csv
dataset=read_csv('Seed_Data.csv')
dataset.head(2)
# data preprocessing 
X=dataset[['P','C','LK','WK']] # input variables 
y=dataset['LKG']  # output variable 
# since y is continuous, therefore we need to apply linear regression
dataset.head(2)
````

F1 = 2 * (precision * recall) / (precision + recall)

# use boston_house_price, model the data using poission regression to predict 'medv'. 

````python
from pandas import read_csv
dataset=read_csv('boston_house_prices.csv')
dataset.head()
y=dataset['MEDV']
X=dataset.drop('MEDV',axis=1)
````

````python
y
````

````python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
model_pois= sm.GLM(y_train, X_train, family=sm.families.Poisson()) # Poisson reg
res_p = model_pois.fit()
res_p.summary()
pred_p=res_p.predict(X_test)
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse_p = sqrt(mean_squared_error(y_test, pred_p))
rmse_p
#linear regression, evaluate the model in terms of rmse and compare the result versus poisson regression 

# linear regression model
model_lr= sm.GLM(y_train, X_train, family=sm.families.Gaussian())  # linear regression 
res_lr=model_lr.fit()
pred_lr=res_lr.predict(X_test)
rmse_lr = sqrt(mean_squared_error(y_test, pred_lr))
print(rmse_p,rmse_lr)
````

````python
 RMSE_p=[]
 RMSE_L=[]
for i in range(100):
  X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
  #poisson regression
  model_pois= sm.GLM(y_train, X_train, family=sm.families.Poisson()) # Poisson reg
  res_p = model_pois.fit()
  pred_p=res_p.predict(X_test)
  rmse_p = sqrt(mean_squared_error(y_test, pred_p))
  RMSE_p.append(rmse_p)
  #Linear regression 
  model_lr= sm.GLM(y_train, X_train, family=sm.families.Gaussian())
  res_lr=model_lr.fit()
  pred_lr=res_lr.predict(X_test)
  rmse_lr = sqrt(mean_squared_error(y_test, pred_lr))
  RMSE_L.append(rmse_lr)

[np.mean(RMSE_p),np.mean(RMSE_L)]
````

# Conclusion:
