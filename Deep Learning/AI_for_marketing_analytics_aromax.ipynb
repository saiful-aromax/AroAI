```python
def get_x_y_and_scaler(symbol,exchange,training_percentage,input_sequence_length):

  apple_index_data = tv.get_hist(symbol=symbol,exchange=exchange,interval=Interval.in_4_hour,n_bars=10000)

  data = apple_index_data.filter(['close'])

  #Convert the dataframe to numpy array
  dataset = data.values

  # Get the number of rows to train the model on
  import math
  training_data_len = math.ceil(len(dataset)*training_percentage) # take training_percentage% of the data and math.ceil used to round up the number of data

  # Initialize Scaler
  from sklearn.preprocessing import MinMaxScaler
  scaler_close = MinMaxScaler()

  # Scaling the data
  scaled_data_close = scaler_close.fit_transform(dataset)

  sc_data_close  = []
  for i in scaled_data_close:
    sc_data_close.append(i.tolist()[0])


  x_train = []
  y_train = []
  x_test = []
  y_test = []

  for i in range(input_sequence_length, training_data_len):
    x_train.append(sc_data_close[i-input_sequence_length:i])
    y_train.append(sc_data_close[i:i+input_sequence_length])

  for i in range(training_data_len,len(sc_data_close)-7):
    x_test.append(sc_data_close[i-input_sequence_length:i])
    y_test.append(sc_data_close[i:i+input_sequence_length])

  x1_train, y1_train = np.array(x_train), np.array(y_train)
  x1_test, y1_test = np.array(x_test), np.array(y_test)

  return x1_train,x1_test,y1_train,y1_test,scaler_close

```

```python
def prepareDataUpperLowerBand(df, time_interval):
  df['MA'] = df['Close'].rolling(window=time_interval).mean()

  # Calculate the i-day standard deviation
  df['STD'] = df['Close'].rolling(window=time_interval).std()

  # Calculate Bollinger Bands
  df['B_U'] = df['MA'] + 2 * df['STD']
  df['B_L'] = df['MA'] - 2 * df['STD']

  df = df[['MA','B_U','B_L','Close']].dropna()
  return df
```
