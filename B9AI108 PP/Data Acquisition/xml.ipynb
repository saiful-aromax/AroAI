```python
product=tree[1]
for e in product[0][0]:
  print(e.tag)
  a=e.attrib
  print(a.get('value',a.get('percent',a.get('mps',a.get('name',0)))))
```

```python
rows=[]
switch=False
for p in product:
  if not switch:
    row={}
  for e in p[0]:
    a=e.attrib
    row[e.tag]=a.get('value',a.get('percent',a.get('mps',a.get('name',0))))
  if switch:
    rows.append(row)
    switch=False
  else:
    rows.append({})
    switch=True
```

```python
pd.json_normalize(rows)
```
